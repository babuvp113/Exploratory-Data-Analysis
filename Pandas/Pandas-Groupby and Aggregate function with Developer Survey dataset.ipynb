The usage of the Social media and Salary of the individual varies and depends on the locality 
#### So here we use the concept of Grouping and Aggregating 
Now we will group Salary based on the location

```python
# Lets do it by filter method without using 'groupby' function
data.loc[data['Country'] == 'India']['SocialMedia'].value_counts()
#Popularity of different social media's in India for the users who participated in the Survey
```

```python
#Popularity of different social media's in United States for the users who participated in the Survey
data.loc[data['Country'] == 'United States']['SocialMedia'].value_counts()
```

```python
# Lets use Groupby function to do the above job
data.groupby(['Country'])['SocialMedia'].value_counts()
```

```python
# From above we can see only 2 countries response, lets find out 50 countries response
data.groupby(['Country'])['SocialMedia'].value_counts().head(50)
```

```python
#Lets find out the results for India and United States whether they are matching with above count which we calculated without Grupby function
data.groupby(['Country'])['SocialMedia'].value_counts().loc[['India','United States']]
```

```python
# If you want to see the results in %
data.groupby(['Country'])['SocialMedia'].value_counts(normalize = True).loc[['India','United States']]
```

```python
#Lets find out median salary in diff location using Groupby function
data.groupby(['Country'])['ConvertedComp'].median().loc[['India','United States','United Kingdom']]
```

```python
# Use of Aggregate function to see Mean and Median Salary
data.groupby(['Country'])['ConvertedComp'].agg(['mean','median'])
```

```python
data.groupby(['Country'])['ConvertedComp'].agg(['mean','median']).loc[['India','United States','Germany','United Kingdom']]
```

```python
data.groupby(['Country'])['ConvertedComp'].mode()
# We can see that mode cant apply to grouby fucntion directly
```

```python
data.groupby(['Country'])['ConvertedComp'].apply(lambda x:x.mode())
```

```python
#Lets see which country uses most popularly which programming Language[By Filter method
data.loc[data['Country'] == 'India']['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
data.loc[data['Country'] == 'United States']['LanguageWorkedWith'].str.contains('C').sum()
```

```python
# Lets apply Grouby function to evaluate the above one
data.groupby(['Country'])['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# SO again we need to apply the Apply function

data.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
```

```python
data.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum()).loc['India']
```

```python
data.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x:x.str.contains('C').sum()).loc[['India','United States']]
```

```python
# Now Lets calculate what is the % of users from each country who knows python
total_respondents = data['Country'].value_counts()
total_respondents.head() # Total number of Respondents from each country
```

```python
respoKnowPython = data.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
respoKnowPython.head()
```

```python
#Lets create the dataframe of people who knows Python and total number of respondents from each country
df_python = pd.concat([total_respondents,respoKnowPython],axis = 'columns',sort = True)
df_python
```

```python
#Now Rename columns with respective names
df_python.rename(columns={'Country':'NumofRespondents','LanguageWorkedWith':'PeopleKnowsPython'},inplace=True)
df_python
```

```python
df_python['PercofPython'] = df_python['PeopleKnowsPython']/df_python['NumofRespondents']*100
```

```python
df_python
```

```python
df_python.sort_values(by = 'PercofPython',ascending = False,inplace=True)
```
